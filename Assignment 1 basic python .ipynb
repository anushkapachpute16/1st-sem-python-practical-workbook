```python
#Q.1) Write a Python Program to Calculate the Average of Numbers in a Given List.
#CODE:
nums = [10, 20, 30, 40, 50]
total = sum(nums)
count = len(nums)
average = total / count
print("Average of numbers:", average)
```

```python
#Q.2) Write a program which accepts 6 integer values and prints "DUPLICATES" if any of the
 #values entered are duplicates otherwise it prints "ALL UNIQUE".
#Example: Let 6 integers are (32, 10, 45, 90, 45, 6) then output "DUPLICATES" to be printed

#CODE:
nums = []
for i in range(6):
    value = int(input("Enter an integer: "))
    nums.append(value)

if len(nums) != len(set(nums)):
    print("DUPLICATES")
else:
    print("ALL UNIQUE")
```

```python
#Q.3) Write a program which accepts an integer value as input and print "Ok" if value is between
    #1 to 50 (both inclusive) otherwise it prints "Out of range".
#CODE:

num = int(input("Enter number: "))
if num in range(1, 51):
    print("Ok")
else:
    print("Out of range")
```

```python
#Q.4) Write a program which finds sum of digits of a number.
 #CODE:
num = int(input("Enter number: "))
sum_digits = 0

while num > 0:
    sum_digits += num % 10
    num //= 10

print("Sum of digits is:", sum_digits)
```

```python
#Q.5 Write a program which prints Fibonacci series of a number.
#CODE:

n = int(input("Enter number of terms: "))
a = 0
b = 1
print("Fibonacci series:")
for i in range(n):
    print(a, end=" ")
    c = a + b 
    a = b
    b = c
```

```python
#Q.1) Write a program which accept an integer value 'n' and display all prime numbers till 'n'
#CODE:

n = int(input("Enter a number: "))

for num in range(2, n + 1):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num, end=" ")
```

```python
#Q.2 Write a program that accept two integer values and if both are equal then prints "SAME
    #identity" otherwise prints "DIFFERENT identity".
#CODE:

a = int(input("Enter first integer: "))
b = int(input("Enter second integer: "))
if a == b:
    print("SAME identity")
else:
    print("DIFFERENT identity")
```

```python
#Q.3) Write a program to display following pattern:
#1 2 3 4
#1 2 3
#1 2
#1

#CODE:

n = 4
for i in range(n, 0, -1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()  
```

```python
#Q.4) Write a program to reverse a given number.

num = int(input("Enter a number: "))
reverse_num = 0
while num > 0:
    digit = num % 10            
    reverse_num = reverse_num * 10 + digit
    num = num // 10              
print("Reversed number is:", reverse_num)
```
