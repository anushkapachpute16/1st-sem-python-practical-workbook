<h4>Question Set A

```python
# Q.1) Write a program to find the union, intersection, and difference of two sets entered by the user
s1 = set(input("Enter first set elements: ").split())
s2 = set(input("Enter second set elements: ").split())

print("Union:", s1 | s2)
print("Intersection:", s1 & s2)
print("Difference:", s1 - s2)
```

```python
# Q.2) Write a program to check if a set is a subset of another set. 
set1 = {1, 2}
set2 = {1, 2, 5, 4}
print(set1.issubset(set2))
print(set2.issubset(set1))
```

```python
# Q.3) Write a program to remove duplicates from a list using sets.
lst=[1,2,8,3,8,4]
print(list(set(lst)))
```

```python
# Q.4) Write a program to find elements that are present in the first set but not in the second set.
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
result = set1 - set2
print(result)
```

```python
# Q.5) Write a program to check if two sets are disjoint (have no common elements).
set1 = {1, 2, 3}
set2 = {4, 5, 6}
set3 = {3, 4, 5}
print(set1.isdisjoint(set2)) 
print(set1.isdisjoint(set3))
```

```python
# Q.1) Write a program that takes two sentences as input and finds: 
# ● Common words in both sentences 
# ● Words unique to the first sentence 
# ● Words unique to the second sentence

s1="this is rohan"; s2="rohan is my best friend"
set1=set(s1.split()); set2=set(s2.split())
print("Common",set1&set2)
print("Unique to first",set1-set2)
print("Unique to second",set2-set1)
```
