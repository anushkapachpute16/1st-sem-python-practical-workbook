```python
#Q.1) Write a program to count the number of vowels in a given string.
string = input("Enter a string: ")
count = 0

for v in "aeiouAEIOU":
    count += string.count(v)

print("Number of vowels:", count)
```

```python
#Q.2)Write a program to check if a string is a palindrome

s = input("Enter string: ")
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#Q.3) Write a program to reverse a string without using built-in reverse functions.
s= input("Enter string: ")
rev = ""
for ch in s:
    rev = ch + rev
print(rev) 
```

```python
#Q,4) Write a program to count the frequency of each character in a string.
s=input('Enter string: ')
freq={}
for ch in s:
    freq[ch]=freq.get(ch,0)+1
print(freq)
```

```python
#Q.5) Write a program to remove all whitespace from a string.
string = input("Enter string: ")
print(string.replace(" ", ""))
```

<H3>Question Set B</H3>

```python
#Q.1) Write a program to find the first non-repeating character in a string.
s = input("Enter string: ")
freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

for ch in s:
    if freq[ch] == 1:
        print("First non-repeating character:", ch)
        break
```

```python
#Q.2) Write a program to check if two strings are anagrams (contain same characters in different order)
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

if sorted(s1) == sorted(s2):
    print("Anagrams")
else:
    print("Not Anagrams")
```

```python
# Q.3) Write a program to convert a string to title case without using the title() method. 
s = input("Enter string: ")
words = s.split()

for w in words:
    print(w[0].upper() + w[1:], end=" ")
```

```python
# Q.4) Write a program to replace all occurrences of a substring with another substring in a given 
a = input("Enter string: ")
old = input("Enter word to replace: ")
new = input("Enter new word: ")

a = a.replace(old, new)

print("Result is:", a)
```

```python
# Q.1) Write a program to find the longest word in a sentence. 
s = input("Enter sentence: ")
words = s.split()

longest = words[0]

for w in words:
    if len(w) > len(longest):
        longest = w

print("Longest word:", longest)
```

```python
# Q.2) Write a program to check if a string contains only digits, only alphabets, or both.

s = input("Enter a string: ")
if s.isdigit():
    print("Only digits")
elif s.isalpha():
    print("Only alphabets")
else:
    print("Both")
```
