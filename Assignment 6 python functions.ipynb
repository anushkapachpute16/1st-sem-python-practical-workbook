```python
# Q.1) Write a function to find the maximum of three numbers. 
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
print("Maximum number is:", maximum(a,b,c))
```

```python
# Q.2) Write a function to check if a number is prime or not.
def is_prime(n):
    if n<2: return False
    for i in range(2,n):
        if n%i==0: return False
    return True
print(is_prime(79))
```

```python
# Q.3) Write a function to calculate the factorial of a number (both iterative and recursive).
def factorial_iterative(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n-1)

num = int(input("Enter a number: "))

print("Factorial (iterative):", factorial_iterative(num))
print("Factorial (recursive):", factorial_recursive(num))
```

```python
# Q.4) Write a function that accepts a list and returns a new list with unique elements.
def unique_list(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique

print(unique_list([1, 2, 2, 3, 4, 4, 5]))
```

```python
# Q.5) Write a function to calculate the sum of all even numbers in a list.
def sum_of_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

print(sum_of_even([1, 2, 3, 4, 5, 6]))
```

```python
# Q.1) Write a function to convert temperature from Celsius to Fahrenheit and vice versa. 
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

print(celsius_to_fahrenheit(25))
print(fahrenheit_to_celsius(77))
```

```python
# Q.2) Write a function that accepts a string and returns the count of vowels and consonants.
def count_vowels_consonants(s):
    vowels = "aeiouAEIOU"
    v_count = 0
    c_count = 0
    
    for char in s:
        if char.isalpha():
            if char in vowels:
                v_count += 1
            else:
                c_count += 1
    return v_count, c_count

text = input("Enter a string: ")
vowels, consonants = count_vowels_consonants(text)

print("Number of vowels:", vowels)
print("Number of consonants:", consonants)

```

```python
# Q.3) Write a function that accepts variable number of arguments and returns their average
def average(*args):
    # Return 0 if no arguments are passed
    if len(args) == 0:
        return 0

    return sum(args) / len(args)

print(average(10, 20, 30))
```

```python
# Q.4) Write a lambda function to find the maximum of three numbers.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
maximum = lambda a, b, c: a if (a >= b and a >= c) else (b if b >= c else c)
print("Maximum number is:", maximum(a,b,c))
```

```python
# Q.1) Write a recursive function to calculate the sum of digits of a number.
def sum_of_digits(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sum_of_digits(n // 10)

num = int(input("Enter a number: "))

print("Sum of digits:", sum_of_digits(num))
```

```python
# Q.2) Write a function that accepts a list of numbers and returns a dictionary with count of even and odd numbers. 
# Example: [1,2,3,4,5] should return {'even': 2, 'odd': 3}
def count_even_odd(nums):
    even = 0
    odd = 0
    for n in nums:
        if n % 2 == 0:
            even += 1
        else:
            odd += 1
    return {'even': even, 'odd': odd}
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

print("Even and Odd count:", count_even_odd(numbers))
```
